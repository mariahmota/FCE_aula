(a) Na linguagem compilada, todo o código é convertido de uma vez para gerar um arquivo executável, desta forma, o código primeiramente é traduzido em formato de máquina antes da execução. Enquanto na linguagem interpretada o código é traduzido linha por linha durante a execução, sendo interpretado e executado ao mesmo tempo.

(b) A programação imperativa é um paradigma que foca na sequência de comandos que devem ser executados, podendo incluir ferramentas como condicionais ou loops para auxiliar no controle do fluxo de execucação. Já a programação procedural, que é um subtipo da imperativa, concentra-se na modularização do código através de funções e/ou procedimentos, de forma que as instruções são realizadas em blocos. Neste sentido, facilita a manutenção e reutilização do código. Portanto, a abordagem procedural, em comparação a imperativa, é mais eficiente e permite uma melhor organização na estruturação.

```python
# Importação de biblioteca
import numpy as np
```

```python
# Definição de variáveis
## Para 20 min de monitoramento, tem-se 10 leituras de pressão
pressao = np.random.uniform(1.0,6.000001,10)
## Faixa de pressão ideal
pressao_min = 2.0
pressao_max = 5.0
```

```python
# Definição da função
def sistema(pressao):

  """
  Esta função simula o monitoramento da pressão do trocador de calor, informando:
  - O número da leitura
  - O valor da pressão
  - Quantos alarmes foram ativados
  - Status do sistema
  Após 3 alarmes o sistema é interrompido, logo, o código não continua a
  execução das leituras
  """
 # Comandos da função
  leitura = 0
  alarme = 0
  status = "Operando"

  for i in pressao:
    leitura = leitura + 1
    if not (pressao_min <= i <= pressao_max):
      alarme = alarme + 1
    if alarme == 3:
      status = "Interrompido"
      print('Leitura %d:' % leitura, 'Pressão = %.1f bar,' % i,'Alarmes ativados = %d/3,' % alarme, 'Sistema =', status)
      break
    else:
      print('Leitura %d:' % leitura, 'Pressão = %.1f bar,' % i,'Alarmes ativados = %d/3,' % alarme, 'Sistema =', status)
```

```python
# Chamada da função
sistema(pressao)
```
